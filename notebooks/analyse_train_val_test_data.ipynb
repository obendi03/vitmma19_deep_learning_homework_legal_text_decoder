```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
```

```python
DATA_DIR = "../data/"

# F√°jlok ellen≈ërz√©se
train_path = os.path.join(DATA_DIR, 'train.csv')
val_path = os.path.join(DATA_DIR, 'val.csv')
test_path = os.path.join(DATA_DIR, 'test.csv')
```

```python
print(f"üîç Ellen≈ërz√©s: {DATA_DIR}")
for path in [train_path, val_path, test_path]:
    exists = os.path.exists(path)
    print(f"  {path}: {'‚úÖ' if exists else '‚ùå'}")
    if not exists:
        print("‚ùå HIBA: F√°jlok hi√°nyoznak!")
        exit()
```

```python
# ========================================
# üìä 3 K√úL√ñN OSZLOP DIAGRAM - EGYENLEGESEN
# ========================================

# Beolvas√°s + NUMERIC KONVERT√ÅL√ÅS
train_df = pd.read_csv(train_path, names=['text', 'rating'])
val_df = pd.read_csv(val_path, names=['text', 'rating'])
test_df = pd.read_csv(test_path, names=['text', 'rating'])

train_df['rating'] = pd.to_numeric(train_df['rating'], errors='coerce')
val_df['rating'] = pd.to_numeric(val_df['rating'], errors='coerce')
test_df['rating'] = pd.to_numeric(test_df['rating'], errors='coerce')

print(f"\nüìä Dataset m√©retek:")
print(f"Train: {len(train_df):,} | Val: {len(val_df):,} | Test: {len(test_df):,}")

colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FFEAA7', '#DDA0DD']
```

```python
# ========================================
# 1Ô∏è‚É£ TRAIN OSZLOP DIAGRAM
# ========================================
train_counts = train_df['rating'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
bars = plt.bar(train_counts.index.astype(int), train_counts.values, 
               color=colors, alpha=0.8, edgecolor='black')
plt.title('üöÇ Train Set - √ârt√©kel√©sek Eloszl√°sa', fontweight='bold', fontsize=16)
plt.xlabel('√ârt√©kel√©s (1-5)', fontweight='bold')
plt.ylabel('Darabsz√°m', fontweight='bold')
plt.ylim(0, max(train_counts.values)*1.1)
plt.grid(axis='y', alpha=0.3)

for bar, value in zip(bars, train_counts.values):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + height*0.01,
             f'{int(value):,}', ha='center', va='bottom', fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# ========================================
# 2Ô∏è‚É£ VAL OSZLOP DIAGRAM
# ========================================
val_counts = val_df['rating'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
bars = plt.bar(val_counts.index.astype(int), val_counts.values, 
               color=colors, alpha=0.8, edgecolor='black')
plt.title('‚úÖ Validation Set - √ârt√©kel√©sek Eloszl√°sa', fontweight='bold', fontsize=16)
plt.xlabel('√ârt√©kel√©s (1-5)', fontweight='bold')
plt.ylabel('Darabsz√°m', fontweight='bold')
plt.ylim(0, max(val_counts.values)*1.1)
plt.grid(axis='y', alpha=0.3)

for bar, value in zip(bars, val_counts.values):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + height*0.01,
             f'{int(value):,}', ha='center', va='bottom', fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# ========================================
# 3Ô∏è‚É£ TEST OSZLOP DIAGRAM
# ========================================
test_counts = test_df['rating'].value_counts().sort_index()
plt.figure(figsize=(10, 6))
bars = plt.bar(test_counts.index.astype(int), test_counts.values, 
               color=colors, alpha=0.8, edgecolor='black')
plt.title('üß™ Test Set - √ârt√©kel√©sek Eloszl√°sa', fontweight='bold', fontsize=16)
plt.xlabel('√ârt√©kel√©s (1-5)', fontweight='bold')
plt.ylabel('Darabsz√°m', fontweight='bold')
plt.ylim(0, max(test_counts.values)*1.1)
plt.grid(axis='y', alpha=0.3)

for bar, value in zip(bars, test_counts.values):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + height*0.01,
             f'{int(value):,}', ha='center', va='bottom', fontweight='bold')
plt.tight_layout()
plt.show()

```
