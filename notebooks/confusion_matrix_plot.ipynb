```python
# ========================================
# 📊 4 CONFUSION MATRIX - PNG MENTÉSSEL
# ========================================

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Mappa létrehozása
OUTPUT_DIR = "./evaluation_plots/"
os.makedirs(OUTPUT_DIR, exist_ok=True)

labels = ['1 ', '2 ', '3 ', '4 ', '5 ']

# Confusion Matrix-ok
lstm_test_cm = np.array([
    [0, 0, 5, 4, 4],
    [0, 2, 8, 16, 9],
    [0, 3, 9, 36, 17],
    [0, 0, 13, 58, 35],
    [0, 0, 1, 26, 61]
])

tfidf_test_cm = np.array([
    [0, 2, 4, 2, 5],
    [0, 1, 16, 14, 4],
    [0, 4, 9, 33, 19],
    [0, 1, 24, 47, 34],
    [0, 0, 8, 24, 56]
])

lstm_consensus_cm = np.array([
    [0, 3, 13, 14, 7],
    [0, 2, 16, 56, 17],
    [0, 10, 7, 87, 46],
    [0, 7, 28, 88, 96],
    [0, 4, 5, 25, 64]
])

tfidf_consensus_cm = np.array([
    [2, 1, 8, 16, 10],
    [1, 7, 18, 47, 18],
    [0, 6, 24, 81, 39],
    [0, 3, 35, 113, 68],
    [0, 1, 12, 24, 61]
])

models = ['LSTM Test\n(Accuracy=42.35%)', 'TF-IDF Test \n(Accuracy=36.81%)', 'LSTM Consensus\n(Accuracy=27.06%)', 'TF-IDF Consensus\n(Accuracy=34.79%)']
cms = [lstm_test_cm, tfidf_test_cm, lstm_consensus_cm, tfidf_consensus_cm]

# ========================================
# 📊 4 CONFUSION MATRIX - HEATMAP + PNG MENTÉS
# ========================================
fig, axes = plt.subplots(2, 2, figsize=(20, 16))
fig.suptitle('ÖSSZES CONFUSION MATRIX ÖSSZEHASONLÍTÁS', fontsize=20, fontweight='bold')

for i, (cm, title) in enumerate(zip(cms, models)):
    row, col = i // 2, i % 2
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=labels, yticklabels=labels, ax=axes[row, col],
                cbar_kws={'label': 'Predikciók száma'}, square=True)
    axes[row, col].set_title(title, fontweight='bold', fontsize=14, pad=20)
    axes[row, col].set_xlabel('Predikció', fontweight='bold')
    axes[row, col].set_ylabel('Valódi', fontweight='bold')

plt.tight_layout()

# 💾 PNG MENTÉS
plt.savefig(f'{OUTPUT_DIR}/confusion_matrices_all.png', dpi=300, bbox_inches='tight')
print(f"✅ KÉP MENTVE: {OUTPUT_DIR}/confusion_matrices_all.png")

plt.show()
```
